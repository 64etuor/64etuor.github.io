## 1. 데이터 개요
- 데이터셋: `website_wata.csv`
- 분석 대상: 웹사이트 방문자의 트래픽 소스별 행동 패턴
- 주요 지표: Page Views, Session Duration, Bounce Rate, Previous Visits, Time on Page, Conversion Rate

---

## 2. 공통적인 패턴 분석

### Page Views (페이지 조회수)
- **중앙값**: 모든 트래픽 소스가 **5.0**으로 동일
- **왜도 (Skewness)**: **0.27~0.46**으로 낮은 값 → 데이터 분포가 안정적
- **해석**: 모든 트래픽 소스에서 페이지 조회 수는 비교적 균일하게 유지됨

### Conversion Rate (전환율)
- **평균**: **0.98~0.99**로 매우 높은 전환율
- **왜도**: **-3.7~-5.9**로 강한 음의 왜도 → 대부분의 사용자가 전환을 성공적으로 수행
- **해석**: 전체적으로 전환율이 높으며, 일부 이상적으로 높은 값이 존재할 가능성이 있음

## 3. 트래픽 소스별 특징

### **Organic (자연 검색)**
**가장 안정적인 체류시간**
   - Session Duration 표준편차(Std): **3.13** (다른 소스보다 변동성이 낮음)  
**낮은 이탈률(Bounce Rate)**
   - 평균 **0.28**, 왜도 **0.681** → 비교적 균형 잡힌 분포  
**Previous Visits 변동성 낮음**
   - 표준편차(Std): **1.42** → 사용자가 반복 방문하는 경향이 있음  
**인사이트**:  
Organic 트래픽은 안정적인 행동 패턴을 보이며, 재방문율이 높은 사용자층이 많음.

---

### **Paid (유료 광고)**
**가장 높은 Session Duration 변동성**
   - Session Duration 표준편차(Std): **3.36** (다른 소스보다 크며, 변동성이 큼)  
**Time on Page 최고값**
   - **최대 24.80분** → 일부 사용자가 페이지에서 오랜 시간 머뭄  
**Session Duration 왜도 높음**
   - **2.275**로 가장 높은 값 → 일부 사용자가 매우 긴 체류시간을 가짐  
**인사이트**:  
Paid 트래픽은 사용자 체류시간의 변동성이 크지만, 광고로 유입된 방문자가 장시간 머물 가능성이 있음.

---

### **Social (소셜 미디어)**
**가장 낮은 Page Views 평균**
   - 평균 **4.70** → 다른 트래픽 소스보다 적은 페이지 조회  
**Conversion Rate 왜도가 가장 큼**
   - 왜도 **-5.934** → 대부분의 사용자가 빠르게 전환하거나 이탈  
**Previous Visits 높은 왜도**
   - 왜도 **1.023** → 일부 사용자는 이전 방문 기록이 많음  
**인사이트**:  
Social 트래픽은 페이지 탐색은 적지만, 충성도 높은 사용자(반복 방문자)들이 존재하며 빠르게 전환하는 경향이 있음.

---

### **Referral (추천 트래픽)**
**가장 낮은 Bounce Rate**
   - 평균 **0.27**, 중앙값 **0.25** → 방문자가 페이지를 이탈하지 않고 머무르는 경향  
**이상치 없음**
   - Page Views와 Bounce Rate에서 이상치(Outlier) 비율 **0.0%**  
**가장 높은 Conversion Rate 평균**
   - 평균 **0.99** → 추천을 통한 유입이 전환에 가장 효과적  
**인사이트**:  
Referral 트래픽은 이탈률이 낮고, 전환율이 가장 높아 **가장 효과적인 유입 채널**로 보임.

## 4. 인사이트 요약

| 트래픽 소스 | 특징 | 주요 인사이트 |
|------------|-----------------|---------------------------|
| **Organic** | 안정적인 체류시간 & 낮은 이탈률 | 재방문율이 높아 지속적 SEO 최적화 필요 |
| **Paid** | 체류시간 변동성 높음 | 광고 타겟팅을 최적화하여 ROI 개선 가능 |
| **Social** | 페이지뷰 낮지만 빠른 전환 | 충성도 높은 유저 대상 리타겟팅 강화 필요 |
| **Referral** | 전환율 최고 & 낮은 이탈률 | 추천 트래픽 증가 전략(파트너십, 제휴) 필요 |

**모든 트래픽 소스에서 전환율이 높으며, 특히 Referral이 가장 효과적**  
**Paid 트래픽은 체류시간이 변동성이 크지만 전환율은 유지됨**  
**Organic 트래픽은 전반적으로 가장 안정적인 행동 패턴을 보임**  
**Social 트래픽은 페이지뷰는 낮지만 충성도 높은 방문자가 존재**  

**트래픽 소스별 추천 전략**  
1. **Referral 트래픽 확대** → 제휴 마케팅, 추천 프로그램 활성화  
2. **Paid 트래픽 최적화** → 광고 타겟팅 개선, 체류시간 높은 사용자 분석  
3. **Organic 지속 개선** → SEO 최적화 및 콘텐츠 강화  
4. **Social 리타겟팅 강화** → 충성도 높은 사용자 대상 캠페인 실행(타겟 마케팅)

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict
from dataclasses import dataclass
````

````python
@dataclass
class VisualizationConfig:
    figure_size: tuple = (15, 10)
    font_family: str = 'Malgun Gothic'
    rotation: int = 45
    correlation_figsize: tuple = (10, 8)
    
    def setup(self):
        plt.rcParams['font.family'] = self.font_family
        plt.rcParams['axes.unicode_minus'] = False

class WebsiteAnalyzer:
    
    def __init__(self, config: VisualizationConfig):
        self.config = config
        self.config.setup()
        self.metrics = [
            'Page Views', 'Session Duration', 'Bounce Rate',
            'Time on Page', 'Previous Visits', 'Conversion Rate'
        ]
    
    def load_data(self, file_path: str) -> pd.DataFrame:
        return pd.read_csv(file_path)
    
    def analyze_traffic_sources(self, df: pd.DataFrame) -> Dict:
        return df.groupby('Traffic Source').agg({
            metric: 'mean' for metric in self.metrics
        }).to_dict('index')
        
    def analyze_metrics(self, df: pd.DataFrame) -> None:
        for source in df['Traffic Source'].unique():
            source_data = df[df['Traffic Source'] == source]
            print(f"\n=== {source} Traffic Analysis ===")
            
            for metric in self.metrics:
                metric_data = source_data[metric]
                
                stats = {
                    'median': metric_data.median(),
                    'mean': metric_data.mean(),
                    'std': metric_data.std(),
                    'min': metric_data.min(),
                    'max': metric_data.max(),
                    'skewness': metric_data.skew()
                }
                
                Q1 = metric_data.quantile(0.25)
                Q3 = metric_data.quantile(0.75)
                IQR = Q3 - Q1
                outlier_count = len(metric_data[(metric_data < Q1 - 1.5 * IQR) | 
                                            (metric_data > Q3 + 1.5 * IQR)])
                outlier_ratio = outlier_count / len(metric_data)
                
                print(f"\n[{metric}]")
                print(f"Median: {stats['median']:.2f}")
                print(f"Mean: {stats['mean']:.2f}")
                print(f"Std: {stats['std']:.2f}")
                print(f"Min: {stats['min']:.2f}")
                print(f"Max: {stats['max']:.2f}")
                print(f"Outlier ratio: {outlier_ratio:.1%}")
                print(f"Skewness: {stats['skewness']:.3f}")

    def plot_boxplots(self, df: pd.DataFrame):
        fig = plt.figure(figsize=self.config.figure_size)
        
        for i, metric in enumerate(self.metrics, 1):
            ax = plt.subplot(2, 3, i)
            sns.boxplot(x='Traffic Source', y=metric, data=df)
            plt.title(f'{metric} by Traffic Source')
            plt.xticks(rotation=self.config.rotation)
        
        plt.tight_layout()

    def plot_violin(self, df: pd.DataFrame):
        plt.figure(figsize=self.config.figure_size)
        
        for i, metric in enumerate(self.metrics, 1):
            plt.subplot(2, 3, i)
            
            sns.violinplot(x='Traffic Source', y=metric, data=df, 
                        inner='box',
                        cut=0)
            
            plt.title(f'{metric} Distribution by Traffic Source')
            plt.xticks(rotation=self.config.rotation)
            
            if metric != 'Conversion Rate':
                q1, q3 = df[metric].quantile([0.25, 0.75])
                iqr = q3 - q1
                plt.ylim(max(q1 - 1.5 * iqr, 0), q3 + 1.5 * iqr)
        
        plt.tight_layout()
        return plt.gcf()
    
    def plot_correlation_heatmap(self, df: pd.DataFrame):
        plt.figure(figsize=self.config.correlation_figsize)
        correlation = df[self.metrics].corr()
        sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
        plt.title('Correlation between Metrics')
        plt.tight_layout()
````

````python
# 데이터 분석 클래스 초기화 및 데이터 로드드

config = VisualizationConfig()
analyzer = WebsiteAnalyzer(config)
df = analyzer.load_data('website_wata.csv')
````

````python
# 분석 결과 출력

analyzer.analyze_metrics(df)
````

````python
# 분석 실행 및 시각화

metrics = analyzer.analyze_traffic_sources(df)

analyzer.plot_boxplots(df)

analyzer.plot_violin(df)

analyzer.plot_correlation_heatmap(df)
````
