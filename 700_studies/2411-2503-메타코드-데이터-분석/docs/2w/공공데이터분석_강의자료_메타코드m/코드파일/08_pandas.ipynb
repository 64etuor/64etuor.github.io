### 출동소요시간

```python
# 구별 평균 출동소요시간
df_seoul['출동소요시간'] = df_seoul['출동일시'] - df_seoul['신고일시']
df_seoul
```

```python
#출동소요시간_초
df_seoul['출동소요시간_초'] = df_seoul['출동소요시간'].dt.total_seconds().astype(int)
df_seoul
```

```python
# 출동소요시간_분
df_seoul['출동소요시간_분'] = df_seoul['출동소요시간'].dt.total_seconds() / 60
df_seoul
```

```python
# df_seoul['출동소요시간_분'].unique().tolist()
df_seoul['출동소요시간_분'] = df_seoul['출동소요시간_분'].astype(int)
df_seoul
```

```python
# describe() : 통계적 요약
pd.set_option('display.float_format', '{:.2f}'.format)
df_seoul.describe(include='number')
```

```python
df_seoul[df_seoul['신고일시']>df_seoul['출동일시']].sort_values(['출동소요시간'], ascending=True)
```

```python
# boxplot
# 데이터 분포, 이상치 시각화
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(20,5))
sns.boxplot(data=df_seoul, x='발생장소_구', y='출동소요시간_분', hue='발생장소_구')
```

```python
df_seoul = df_seoul[df_seoul['신고일시']<df_seoul['출동일시']]
plt.figure(figsize=(20,5))
sns.boxplot(data=df_seoul, x='발생장소_구', y='출동소요시간_분', hue='발생장소_구')
```

```python
df_seoul['출동소요시간_분'].describe()
```

```python
# 평균 +- 표준편차 * 3
mean = df_seoul['출동소요시간_분'].mean()
std = df_seoul['출동소요시간_분'].std()
threshold = mean + 3 * std #평균에서 표준편차의 3배 수준 이상이면 이상치로 판단

df_cleaned = df_seoul[df_seoul['출동소요시간_분']<=threshold]
df_cleaned
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(data=df_cleaned, x='발생장소_구', y='출동소요시간_분', hue='발생장소_구')
```

```python
df_cleaned_pivot = df_cleaned.pivot_table(index=['발생장소_구'],values='출동소요시간_분',aggfunc='mean').reset_index()
df_cleaned_pivot
```

```python
plt.figure(figsize=(20,5))
sns.barplot(data=df_cleaned_pivot, x='발생장소_구', y='출동소요시간_분', hue='발생장소_구')
```

```python
mean_time = df_cleaned_pivot['출동소요시간_분'].mean()

plt.figure(figsize=(20,5))
sns.barplot(data=df_cleaned_pivot, x='발생장소_구', y='출동소요시간_분', hue='발생장소_구')

# mean_time을 선으로 추가
plt.axhline(mean_time, color='black', linestyle='--', label='평균 출동 소요시간')
plt.legend()

# 그래프 출력
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
df_fire
```

```python
df_fire.iloc[0]
```

```python
center = df_fire.iloc[0]

# 지도를 생성
map = folium.Map(location=[center['위도'], center['경도']], zoom_start=13)

folium.Marker(
    location=[center['위도'], center['경도']],
    popup=center['서소이름'],
).add_to(map)

# 지도 출력
map
```

```python
center = df_fire.iloc[0]

# 지도를 생성
map = folium.Map(location=[center['위도'], center['경도']], zoom_start=11)

# 모든 소방서 위치에 마커를 추가
for _, row in df_fire.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['서소이름'],
    ).add_to(map)

# 지도 출력
map
```

```python
center = df_fire.iloc[0]

# 지도를 생성
map = folium.Map(location=[center['위도'], center['경도']], zoom_start=11)

# 모든 소방서 위치에 마커를 추가
for _, row in df_fire.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        icon=folium.Icon(color='red', icon='info-sign'),
        popup=folium.Popup(f"<h5>{row['서소이름']}</h5>", max_width=300),
    ).add_to(map)

# 지도 출력
map
```
