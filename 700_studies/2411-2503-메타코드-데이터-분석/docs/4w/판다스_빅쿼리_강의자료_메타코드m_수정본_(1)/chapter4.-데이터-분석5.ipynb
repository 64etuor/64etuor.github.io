```python
location_data = users[['id','latitude','longitude']]
```

```python
using_data = users[users['country']=='South Korea'][['latitude','longitude']].drop_duplicates()
```

```python
using_data.head()
```

```python
# 지도 띄우기, 부산 중심
map = folium.Map(location = [35.1796, 129.0756], zoom_start=11)
```

```python
map
```

```python
for i, j in using_data.values[:100]:
    popup = folium.Popup('1', max_width=200)
    folium.Marker(location=[i, j], popup=popup).add_to(map)
map
```
